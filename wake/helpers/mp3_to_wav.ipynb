```python
from pydub import AudioSegment
import os
from glob import glob
import shutil
# Function to convert audio files to WAV format
def convert_to_wav(path, save_path):
    folder=path
    waves = glob(folder+'/'+ '*.mp3')
    for i in waves:
        w = i
        try:
            # Load the audio file
            audio = AudioSegment.from_file(i)

            audio_name = w.split("\\")[-1].split('.')[0] + ".wav"
            print (audio_name)
            wav_path = os.path.join(save_path, audio_name)
            # Export the audio in WAV format
            audio.export(wav_path, format="wav")

            print(f"Conversion successful: {i} -> {audio_name}")

        except Exception as e:
            print(f"Conversion failed for {i}: {e}")
```

```python
def pad_clip_audio(path, sr, duration_ms, save_path):
    folder=path
    waves = glob(folder+'/'+ '*.wav')
    # Check if the folder path exists
    if not os.path.exists(save_path):
        # If it doesn't exist, create the folder
        os.makedirs(save_path)
        print(f"Folder '{save_path}' created.")
    for i in waves:
        w = i
        try:
            silent_audio = AudioSegment.silent(duration=duration_ms)
            audio = silent_audio.overlay(AudioSegment.from_wav(i))
            audio_name = w.split("\\")[-1]
            print (audio_name)
            wav_path = os.path.join(save_path, audio_name)
            # Export the audio in WAV format
            audio.export(wav_path, format="wav")

            print(f"Conversion successful: {i} -> {audio_name}")
            #audio = np.frombuffer(audio.set_frame_rate(sr).set_channels(1)[0:duration_ms]._data, dtype='int16')
        except Exception as e:
            print(f"Conversion failed for {i}: {e}")
```
